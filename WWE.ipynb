# Creating charts

```python
max_reign = wwe[wwe['Ruled'] == wwe.groupby('Belt Name')['Ruled'].transform('max')]
max_reign = max_reign[['Belt Name', 'Ruled']].reset_index(drop=True)
colors = ['red', 'yellow',] * (len(max_reign) // 2) + ['red'] * (len(max_reign) % 2)
ax = max_reign.plot.barh(x='Belt Name', y='Ruled', legend=False, color=colors)
ax.bar_label(ax.containers[0])
plt.xlabel('Ruled')
plt.ylabel('Belt Name')
plt.title('Maximum Ruled per Belt Name')
plt.show()
```

```python
max_days = wwe[wwe['No of days'] == wwe.groupby('Belt Name')['No of days'].transform('max')]
max_days = max_days[['Belt Name', 'No of days']].reset_index(drop=True)
colors = ['red', 'yellow',"blue"] * (len(max_days) // 2) + ['yellow'] * (len(max_days) % 2)
ax = max_days.plot.pie(y='No of days', labels=max_days['Belt Name'], autopct='%1.1f%%', startangle=90, colors=colors, figsize=(10, 10))
ax.set_ylabel('')
plt.title('Maximum Number of Days per Belt Name')
plt.show()
```

```python
mask_wwe_championship = wwe['Belt Name'] == 'WWE Championship'
year_group = wwe[mask_wwe_championship].groupby(wwe['Date'].dt.year)['Ruled'].count()
year_group_mean = year_group.mean()
ax = year_group.plot.bar(ylabel='WWE Championship Title Changes', xlabel='Year')
ax.bar_label(ax.containers[0])
plt.axhline(y=year_group_mean, color='b', linestyle='--')
plt.text(-0.5, year_group_mean + 0.5, 'average = {:.1f}'.format(year_group_mean), color='b')
plt.show()
```

```python
city_group = wwe[mask_wwe_championship].groupby('Event location')['Ruled'].count().sort_values()
ax = city_group.plot.barh(figsize=(15,30), xlabel='Number of Wins')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
event_group = wwe[mask_wwe_championship].groupby('Event Name')['Ruled'].count().sort_values()
ax = event_group.plot.barh(figsize=(15,30), xlabel='Number of Wins')
ax.bar_label(ax.containers[0])
plt.show()
```
