# Matplotlib

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Single line polt

```python
month=np.arange(1,13)
month
```

```python
mouseprice=np.random.randint(10,50,12)
mouseprice
```

```python
plt.plot(month,mouseprice,color="green",marker="*")
plt.grid()
plt.xlabel("month")
plt.ylabel("price")
plt.title("Monthly trend of price")
plt.show()
```

# multiple line plot

```python
laptoppricee=np.random.randint(20,90,12)
laptoppricee
```

```python
plt.plot(month,mouseprice,color="red",marker="*",label="mouse")
plt.plot(month,laptoppricee,color="gold",marker="d",label="laptop")
plt.grid()
plt.xlabel("month")
plt.ylabel("price")
plt.title("Monthly trend of price")
plt.legend()
plt.show()
```

# scatter plot

```python
df=sns.load_dataset("iris")
df.head(7)
```

```python
plt.scatter(x=df['sepal_length'],y=df['petal_length'],color='green',marker='*')
plt.title("sepal length vs petal length")
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.grid()
plt.show()
```

#### Inference:-
* As the sepal length increases, the petal length also increases. Thus, it shows positive correlation

```python
plt.scatter(x=df['sepal_length'],y=df['sepal_width'],color='green',marker='*',label="sepalwidth")
plt.scatter(x=df['sepal_length'],y=df['petal_length'],color='blue',marker='d',label='petallength')
plt.title("sepal length vs petal length")
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.legend()
plt.grid()
plt.show()
```

# bar plot


```python
name=["kams","axel","alan","raj"]
mark=[20,45,57,56]
plt.bar(x=name,height=mark,color=["green","blue","red","orange"],label=name)
plt.title("marks analysis")
plt.xlabel("name")
plt.ylabel("marks")
plt.legend()
plt.show()
```

# horizontal bar plot

```python
name=["kams","axel","alan","raj"]
mark=[20,45,57,56]
plt.barh(y=name,width=mark,color=["green","blue","red","orange"],label=name)
plt.title("marks analysis")
plt.xlabel("marks")
plt.ylabel("name")
plt.legend()
plt.show()
```

# box plot

```python
df.head()
```

```python
df['age'].isnull().sum()
```

```python
df['age'].shape
```

```python
age = df['age'].dropna()
```

```python
age.describe()
```

```python
plt.boxplot(age)
plt.text(x=1.1, y = age.min(), s = "min")
plt.text(x=1.1, y = age.quantile(0.25), s = "Q1")
plt.text(x=1.1, y = age.median(), s = "median(Q2)")
plt.text(x=1.1, y = age.quantile(0.75), s = "Q3")
plt.text(x=1.1, y = age.max(), s = "max")
plt.title("Boxplot of Age")
plt.xlabel("Age")
plt.ylabel("Values")
plt.show()
```
