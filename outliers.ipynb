```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
wwe=pd.read_csv("data.csv")
```

```python
wwe
```

```python
wwe.describe()
```

```python
q1=wwe["reign"].quantile(0.25)
q3=wwe["reign"].quantile(0.75)
print(q1)
print(q3)
```

```python
iqr = q3-q1
iqr
```

```python
lower=q1-1.5*iqr
upper=q3+1.5*iqr
print(upper)
print(lower)
```

```python
outliers=wwe[((wwe['reign']<lower)|(wwe['reign']>upper))]
print(outliers)
```

```python
nooutliers=wwe[~((wwe['reign']<lower)|(wwe['reign']>upper))]
nooutliers
```

```python
plt.figure(figsize=(8,5))
sns.distplot(nooutliers['reign'])
plt.title("With outliers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.distplot(outliers['reign'])
plt.title("With out outliers")
plt.show()
```
